```python
import time
import rdflib
from rdflib.graph import Graph, ConjunctiveGraph
import rdflib.namespace
from rdflib.namespace import OWL, RDF, RDFS
from rdflib import Namespace, URIRef, Literal

print ('STARTING')
start = time.time()

lineCounter = 0
lineProgress = 10000

#s = "OpenCyc/opencyc-latest_sample.owl"
s = "OpenCyc/opencyc-latest.owl"

rdfType = URIRef("http://www.w3.org/1999/02/22-rdf-syntax-ns#type")
owlClass = URIRef('http://www.w3.org/2002/07/owl#Class')

allClasses = set()

try:
    g = rdflib.ConjunctiveGraph()
    print ('START PARSING GRAPH')
    g = g.parse(s)
    print ('DONE PARSING GRAPH')
    #g.serialize(s)
    #print len(g)

    #query = "SELECT ?class WHERE {?class rdf:type owl:Class}"

    #for s, p, o in g.triples((None, None, None)):
        #if (p.eq('http://www.w3.org/1999/02/22-rdf-syntax-ns#type')):
    #    print o

    #for owlClass in g.subjects(RDF.type, OWL.Class):
    #    print owlClass
    
    print ('START READING TRIPLES')
    for s in g.subjects(predicate=rdfType, object=owlClass):
        if s not in allClasses:
            allClasses.add(s)
        lineCounter += 1
        if (lineCounter % lineProgress == 0):
            print ('{} lines read'.format(lineCounter))
    print ('DONE READING TRIPLES')
    print len(allClasses)
    end = time.time()
    print ('EXECUTION TIME: {} s'.format(end-start))
except:
    print ('ERROR')
```

```python
# get top 10 classes
import operator

print ('STARTING')
start = time.time()

lineCounter = 0
lineProgress = 10000

classCount = {} #key: class, value: count
seenClasses = set()

#for s in g.subjects(predicate=rdfType, None):
for s, p, o in g.triples((None, rdfType, None)):
    if s in allClasses:
        if s in seenClasses:
            classCount[s] += 1
        else:
            seenClasses.add(s)
            classCount[s] = 1
print (len(classCount))
top10dict = dict(sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)[:10])
print (sorted(top10dict.items(), key=operator.itemgetter(1), reverse=True))
end = time.time()
print ('EXECUTION TIME: {} s'.format(end-start))
```

```python
# calculate class indegree & outdegree

import operator
import time

print ('STARTING')
start = time.time()

indegreeDic = {} # key:class, value: indegree
outdegreeDic = {} #key: class, value: outdegree

for s, p, o in g.triples((None, None, None)):
    if top10dict.has_key(s):
        if outdegreeDic.has_key(s):
            outdegreeDic[s] += 1
        else:
            outdegreeDic[s] = 1
    if top10dict.has_key(o):
        if indegreeDic.has_key(o):
            indegreeDic[o] += 1
        else:
            indegreeDic[o] = 1

print ('DONE')
print ('INDEGREE RESULT:')
print sorted(indegreeDic.items(), key=operator.itemgetter(1), reverse=True)
print ('OUTDEGREE RESULT')
print sorted(outdegreeDic.items(), key=operator.itemgetter(1), reverse=True)
print ('EXECUTION TIME: {} s'.format(time.time()-start))
```

```python
# get all class instances

import numpy
import operator
import time

print ('STARTING')
start = time.time()

class ClassInstances:
    def __init__(self, name, uri):
        self.name = name
        self.uri = uri
        self.allClassInstances = set()
        self.countDict = dict()
        self.min = 9999999
        self.max = 0
        self.avg = 0.0
        self.median = 0.0
    def getURI(self):
        return self.uri
    def getClassInstances(self):
        return len(self.allClassInstances)
    def addInstance(self, instance):
        if instance in self.allClassInstances:
            self.countDict[instance] += 1
        else:
            self.countDict[instance] = 1
            self.allClassInstances.add(instance)
    def calculateDegrees(self):
        allValueList = []
        for k,v in self.countDict.iteritems():
            allValueList.append(v)
            if (v < self.min):
                self.min = v
            if (v > self.max):
                self.max = v
            self.avg += v
        self.avg = self.avg / len(self.countDict)
        self.median = numpy.median(numpy.array(allValueList))
    def printResults(self):
        print ('{}'.format(self.name))
        print ('min: {}, avg: {}, median: {}, max: {}'.format(self.min, self.avg, self.median, self.max))
        
allInstances = set()
instancesOrg = set()
instancesCharSeq = set()
instancesEvent = set()
instancesInteger = set()
instancesPosInteger = set()
instancesSpaceSur = set()
instancesDevice = set()
instancesPerson = set()
instancesContext = set()
instancesHpo4 = set()

# instantiation of all class instances for indegree          
orgCi = ClassInstances('class instances indegree for orgainzation', 'http://sw.opencyc.org/concept/Mx4rvVjVT5wpEbGdrcN5Y29ycA')
charSeqCi = ClassInstances('class instances indegree for character sequence', 'http://sw.opencyc.org/concept/Mx4rvWEA25wpEbGdrcN5Y29ycA')
eventCi = ClassInstances('class instances indegree for event', 'http://sw.opencyc.org/concept/Mx4rvViADZwpEbGdrcN5Y29ycA')
integerCi = ClassInstances('class instances indegree for integer', 'http://sw.opencyc.org/concept/Mx4rvVieEpwpEbGdrcN5Y29ycA')
posIntegerCi = ClassInstances('class instances indegree for positive integer', 'http://sw.opencyc.org/concept/Mx4rvVie2ZwpEbGdrcN5Y29ycA')
spaceSurCi = ClassInstances('class instances indegree for space surface', 'http://sw.opencyc.org/concept/Mx4rvg1ZeJwpEbGdrcN5Y29ycA')
deviceCi = ClassInstances('class instances indegree for device', 'http://sw.opencyc.org/concept/Mx4rKqORBnS8EdaAAACgyZzFrg')
personCi = ClassInstances('class instances indegree for person', 'http://sw.opencyc.org/concept/Mx4rvViAkpwpEbGdrcN5Y29ycA')
contextCi = ClassInstances('class instances indegree for context', 'http://sw.opencyc.org/concept/Mx4rvViA1ZwpEbGdrcN5Y29ycA')
hpo4Ci = ClassInstances('class instances indegree for HP=4', 'http://sw.opencyc.org/concept/Mx8Ngh4rwBhyv5wpEbGdrcN5Y29ycB4r6dRGtPZPQdady5-2_8lO5A')

# list with references to all indegree classes
classInstancesIndegree = [orgCi, charSeqCi, eventCi, integerCi, posIntegerCi, spaceSurCi, deviceCi, personCi, contextCi, hpo4Ci]

# instantiation of all class instances for outdegree          
orgCo = ClassInstances('class instances outdegree for orgainzation', 'http://sw.opencyc.org/concept/Mx4rvVjVT5wpEbGdrcN5Y29ycA')
charSeqCo = ClassInstances('class instances outdegree for character sequence', 'http://sw.opencyc.org/concept/Mx4rvWEA25wpEbGdrcN5Y29ycA')
eventCo = ClassInstances('class instances outdegree for event', 'http://sw.opencyc.org/concept/Mx4rvViADZwpEbGdrcN5Y29ycA')
integerCo = ClassInstances('class instances outdegree for integer', 'http://sw.opencyc.org/concept/Mx4rvVieEpwpEbGdrcN5Y29ycA')
posIntegerCo = ClassInstances('class instances outdegree for positive integer', 'http://sw.opencyc.org/concept/Mx4rvVie2ZwpEbGdrcN5Y29ycA')
spaceSurCo = ClassInstances('class instances outdegree for space surface', 'http://sw.opencyc.org/concept/Mx4rvg1ZeJwpEbGdrcN5Y29ycA')
deviceCo = ClassInstances('class instances outdegree for device', 'http://sw.opencyc.org/concept/Mx4rKqORBnS8EdaAAACgyZzFrg')
personCo = ClassInstances('class instances outdegree for person', 'http://sw.opencyc.org/concept/Mx4rvViAkpwpEbGdrcN5Y29ycA')
contextCo = ClassInstances('class instances outdegree for context', 'http://sw.opencyc.org/concept/Mx4rvViA1ZwpEbGdrcN5Y29ycA')
hpo4Co = ClassInstances('class instances outdegree for HP=4', 'http://sw.opencyc.org/concept/Mx8Ngh4rwBhyv5wpEbGdrcN5Y29ycB4r6dRGtPZPQdady5-2_8lO5A')
# list with references to all outdegree classes
classInstancesOutdegree = [orgCo, charSeqCo, eventCo, integerCo, posIntegerCo, spaceSurCo, deviceCo, personCo, contextCo, hpo4Co]

def getInstances(s, p, o):
    classString = o.toPython()
    if top10dict.has_key(o):
        allInstances.add(s)
        if (classString == orgCi.getURI()):
            if s not in instancesOrg:
                instancesOrg.add(s)
                return
        if (classString == charSeqCi.getURI()):
            if s not in instancesCharSeq:
                instancesCharSeq.add(s)
                return
        if (classString == eventCi.getURI()):
            if s not in instancesEvent:
                instancesEvent.add(s)
                return
        if (classString == integerCi.getURI()):
            if s not in instancesInteger:
                instancesInteger.add(s)
                return
        if (classString == posIntegerCi.getURI()):
            if s not in instancesPosInteger:
                instancesPosInteger.add(s)
                return
        if (classString == spaceSurCi.getURI()):
            if s not in instancesSpaceSur:
                instancesSpaceSur.add(s)
                return
        if (classString == deviceCi.getURI()):
            if s not in instancesDevice:
                instancesDevice.add(s)
                return
        if (classString == personCi.getURI()):
            if s not in instancesPerson:
                instancesPerson.add(s)
                return
        if (classString == contextCi.getURI()):
            if s not in instancesContext:
                instancesContext.add(s)
                return
        if (classString == hpo4Ci.getURI()):
            if s not in instancesHpo4:
                instancesHpo4.add(s)
                return
        
def countInstanceDegrees(s,p,o):
    if s in allInstances: # outdegree
        #print ('outdegree match')
        if s in instancesOrg:
            orgCo.addInstance(s)
        elif s in instancesCharSeq:
            charSeqCo.addInstance(s)
        elif s in instancesEvent:
            eventCo.addInstance(s)
        elif s in instancesInteger:
            integerCo.addInstance(s)
        elif s in instancesPosInteger:
            posIntegerCo.addInstance(s)
        elif s in instancesSpaceSur:
            spaceSurCo.addInstance(s)
        elif s in instancesDevice:
            deviceCo.addInstance(s)
        elif s in instancesPerson:
            personCo.addInstance(s)
        elif s in instancesContext:
            contextCo.addInstance(s)
        elif s in instancesHpo4:
            hpo4Co.addInstance(s)
    if o in allInstances: #indegree
        #print ('indegree match')
        if o in instancesOrg:
            orgCi.addInstance(o)
        elif o in instancesCharSeq:
            charSeqCi.addInstance(o)
        elif o in instancesEvent:
            eventCi.addInstance(o)
        elif o in instancesInteger:
            integerCi.addInstance(o)
        elif o in instancesPosInteger:
            posIntegerCi.addInstance(o)    
        elif o in instancesSpaceSur:
            spaceSurCi.addInstance(o)
        elif o in instancesDevice:
            deviceCi.addInstance(o)
        elif o in instancesPerson:
            personCi.addInstance(o)
        elif o in instancesContext:
            contextCi.addInstance(o)
        elif o in instancesHpo4:
            hpo4Ci.addInstance(o)

def calculateClassDegrees():
    for item in classInstancesIndegree:
        item.calculateDegrees()
    for item in classInstancesOutdegree:
        item.calculateDegrees()
        
def printClassDegreeResults():
    for item in classInstancesIndegree:
        item.printResults()
    for item in classInstancesOutdegree:
        item.printResults()          
            
#counter = 0
#s, p, o: instance rdf:type top10class
for s, p, o in g.triples((None, rdfType, None)):
    getInstances(s,p,o)      
    #counter += 1
    #if counter > 50000:
    #    break
#print (len(instancesOrg))
#print (len(instancesPerson))
#print (len(allInstances))

counter = 0
for s, p, o in g.triples((None, rdfType, None)):
    countInstanceDegrees(s,p,o)      
    #counter += 1
    #if counter > 50000:
    #    break
calculateClassDegrees()
printClassDegreeResults()

#print (personCi.getClassInstances())
#print (personCo.getClassInstances())
print ('DONE')

print ('EXECUTION TIME: {} s'.format(time.time()-start))
```
