```python
# check how many entities have the "generalizations" relation & 
# check if they are the same that have the categories for entity attribute
import os
import time
import string
import itertools
import re
import operator

start = time.time()

nellSample = 'NELL/NELL_sample.csv'
nell = 'NELL/NELL.08m.995.esv.csv'
readNell = nell


countCat = {} #key: value, value: count
countCat['hasGenRel'] = 0
countCat['noGenRel'] = 0

allEntities = set()
hasGeneralization = set()

lineCounter = 0
lineProgress = 500000

def checkWords(splittedLine):
    word_position = 0
    #newEntity = False
    for word in splitted_line:
        if (word_position==0): #Entity
            entity = word
            if entity not in allEntities:
                # add to list of all entities
                allEntities.add(entity)                
        if (word_position==1): #relation
            relation = word
            if (relation == "generalizations"):
                hasGeneralization.add(entity)
            else:
                return
        word_position += 1
    

print('START')
try:
    # read nell
    fn = open(readNell, 'r')
    for line in fn:
        if (lineCounter != 0):
            splitted_line = line.rstrip('\n').split('\t')
            checkWords(splitted_line)
        lineCounter += 1
        if (lineCounter % lineProgress == 0):
            print ('{} lines read'.format(lineCounter))
    print ('len(allEntities): {}'.format(len(allEntities)))
    print ('len(hasGeneralization): {}'.format(len(hasGeneralization)))
    print ('len(hasCat): {}'.format(len(hasCat)))
    #find different items in both sets
    print('# of different items in hasGeneralization and hasCat: {}'.format(len(hasGeneralization-hasCat)))
    fn.close()
    print ('DONE')
except:
    print('ERROR')

end = time.time()
print ('EXECUTION TIME: {} s'.format(end-start))

```
