```python
# check how many entities have the "categories for entity" attribute
import os
import time
import string
import itertools
import re
import operator

start = time.time()

nellSample = 'NELL/NELL_sample.csv'
nell = 'NELL/NELL.08m.995.esv.csv'
readNell = nell


countCat = {} #key: value, value: count
countCat['hasCat'] = 0
countCat['noCat'] = 0

hasCat = set()
seenEntities = set()

lineCounter = 0
lineProgress = 500000

def checkWords(splittedLine):
    word_position = 0
    #newEntity = False
    for word in splitted_line:
        if (word_position==0): #Entity
            entity = word
        if (word_position==10): # categories for entity
            categories = word
            if entity in seenEntities:
                return
            else:
                seenEntities.add(entity)
                if (categories == ''):
                    countCat['noCat'] += 1
                    return
                else:
                    countCat['hasCat'] += 1
                    hasCat.add(entity)
                    #print ('empty cat')
                    return # break function
        word_position += 1
    

print('START')
try:
    # read nell
    fn = open(readNell, 'r')
    for line in fn:
        if (lineCounter != 0):
            splitted_line = line.rstrip('\n').split('\t')
            checkWords(splitted_line)
        lineCounter += 1
        if (lineCounter % lineProgress == 0):
            print ('{} lines read'.format(lineCounter))
    print countCat
    print ('len(hasCat): {}'.format(len(hasCat)))
    fn.close()
    print ('DONE')
except:
    print('ERROR')

end = time.time()
print ('EXECUTION TIME: {} s'.format(end-start))

```

```python
# check how many entities have the "generalizations" relation
import os
import time
import string
import itertools
import re
import operator

start = time.time()

nellSample = 'NELL/NELL_sample.csv'
nell = 'NELL/NELL.08m.995.esv.csv'
readNell = nell


countCat = {} #key: value, value: count
countCat['hasGenRel'] = 0
countCat['noGenRel'] = 0

allEntities = set()
hasGeneralization = set()

lineCounter = 0
lineProgress = 500000

def checkWords(splittedLine):
    word_position = 0
    #newEntity = False
    for word in splitted_line:
        if (word_position==0): #Entity
            entity = word
            if entity not in allEntities:
                # add to list of all entities
                allEntities.add(entity)                
        if (word_position==1): #relation
            relation = word
            if (relation == "generalizations"):
                hasGeneralization.add(entity)
            else:
                return
        word_position += 1
    

print('START')
try:
    # read nell
    fn = open(readNell, 'r')
    for line in fn:
        if (lineCounter != 0):
            splitted_line = line.rstrip('\n').split('\t')
            checkWords(splitted_line)
        lineCounter += 1
        if (lineCounter % lineProgress == 0):
            print ('{} lines read'.format(lineCounter))
    print ('len(allEntities): {}'.format(len(allEntities)))
    print ('len(hasGeneralization): {}'.format(len(hasGeneralization)))
    print ('len(hasCat): {}'.format(len(hasCat)))
    #find different items in both sets
    print('# of different items in hasGeneralization and hasCat: {}'.format(len(hasGeneralization-hasCat)))
    fn.close()
    print ('DONE')
except:
    print('ERROR')

end = time.time()
print ('EXECUTION TIME: {} s'.format(end-start))

```

```python
import os
import time
import string
import itertools
import re
import operator

start = time.time()

nellSample = 'NELL/NELL_sample.csv'
nell = 'NELL/NELL.08m.995.esv.csv'
readNell = nell

outputFile = 'NELL/generalization.txt'

#header of files: Entity	Relation	Value	Iteration of Promotion	Probability	Source	Entity literalStrings	Value literalStrings	Best Entity literalString	Best Value literalString	Categories for Entity	Categories for Value	Candidate Source


categoryCount = {} #key: category, value: count
genValueCount = {} #key: value, value: count
seenEntities = set()

lineCounter = 0
lineProgress = 500000

def checkWords(splittedLine):
    word_position = 0
    #newEntity = False
    for word in splitted_line:
        if (word_position==0): #Entity
            entity = word
        if (word_position==4):
            prob = float(word)
            if (prob < 1):
                return
        if (word_position==10): # categories for entity
            categories = word
            if (categories == ''):
                #print ('empty cat')
                return # break function
            else:
                if entity in seenEntities:
                    #print('seen: {}'.format(entity))
                    return #break function
                else:
                    #print('not seen: {}'.format(entity))
                    seenEntities.add(entity)
                    for category in categories.split(): #split on whitespace for multiple categories
                        if category not in categoryCount:
                            categoryCount[category] = 0
                        else:
                            categoryCount[category] += 1
#        if (word_position==4): #Probability
#            prob = float(word)
#            if prob==1:
#                if catCandidate not in genValueCount:
#                    genValueCount[catCandidate] = 0
#                else:
#                    genValueCount[catCandidate] += 1
        word_position += 1
    

print('START')
try:
    # read nell
    fn = open(readNell, 'r')
    for line in fn:
        if (lineCounter != 0):
            splitted_line = line.rstrip('\n').split('\t')
            checkWords(splitted_line)
        lineCounter += 1
        if (lineCounter % lineProgress == 0):
            print ('{} lines read'.format(lineCounter))
            #print sorted(categoryCount.items(), key=operator.itemgetter(1), reverse=True)
    print sorted(categoryCount.items(), key=operator.itemgetter(1), reverse=True)
    fn.close()
    print ('DONE')
except:
    print('ERROR')

end = time.time()
print ('EXECUTION TIME: {} s'.format(end-start))

```
